```python
def create_md_dataframe():    
    return pd.DataFrame(columns=list(['MD_name','frame','traj_time','system']))
```

```python
def append_metadata(traj_note, system, location):
        rep_data = []
        try:
            traj = mda.Universe(location + '/md.tpr',
                                location + '/md.xtc')         
            md_name = traj_note
            n_frames = traj.trajectory.n_frames
            ts = traj.trajectory.dt
            for i in range(n_frames):
                rep_data.append([md_name, i, ts * i, system])
                
        except:
            print(traj_note + ' not found.')
        return rep_data
    
    
meta_data = Parallel(n_jobs=num_cores)(delayed(append_metadata)(traj_note = traj_note_dic['traj_note'][i], 
                                                        system = i,
                                                        location = location
                                               )
                           for i, location in enumerate(traj_note_dic['load_location']))
md_data = create_md_dataframe()
for i in range(0,len(traj_note_dic['load_location'])):
    md_data = md_data.append(pd.DataFrame(meta_data[i],columns=list(['MD_name','frame','traj_time','system'])),ignore_index=True)
md_data['frame'] =md_data['frame'].apply(int)
md_data['traj_time'] =md_data['traj_time'].apply(float)
md_data['system'] = md_data['system'].apply(int)
```

```python
from MDAnalysis.analysis import rdf

def append_rdf_to_cation(traj_note, system, location):
    rdf_data = []
    u = mda.Universe(location + '/md.tpr',
                        location + '/md.xtc')
    
    if traj_note == 'CAM':
        cation = u.select_atoms('resname ' + traj_note + ' and type Dz')
    elif traj_note == 'CAL_old':
        cation = u.select_atoms('resname ' + 'CAL')
    else:
        cation = u.select_atoms('resname ' + traj_note)
    water = u.select_atoms('resname TIP3 and name OH2')

    irdf = rdf.InterRDF(cation, water,
                    nbins=75,  # default
                    range=(0.0, 15.0),  # distance in angstroms
                   )
    irdf.run(start=2000,
             stop=md_data[md_data.system==system]['frame'].max() + 1)
    rdf_data.append([irdf.rdf] * (md_data[md_data.system==system]['frame'].max() + 1))
    
    return rdf_data

rdf_data = Parallel(n_jobs=num_cores)(delayed(append_rdf_to_cation)(traj_note = traj_note_dic['traj_note'][i], 
                                                        system = i,
                                                        location = location
                                               )
                           for i, location in enumerate(traj_note_dic['load_location']))

md_data['rdf_cation']  = list(np.concatenate(rdf_data,axis=1)[0])
```

```python
from MDAnalysis.analysis import rdf

def append_rdf_to_anion(traj_note, system, location):
    rdf_data = []
    u = mda.Universe(location + '/md.tpr',
                        location + '/md.xtc')
    
    anion = u.select_atoms('resname CLA')
    water = u.select_atoms('resname TIP3 and name OH2')

    irdf = rdf.InterRDF(anion, water,
                    nbins=75,  # default
                    range=(0.0, 15.0),  # distance in angstroms
                   )
    irdf.run(start=2000,
             stop=md_data[md_data.system==system]['frame'].max() + 1)
    rdf_data.append([irdf.rdf] * (md_data[md_data.system==system]['frame'].max() + 1))
    
    return rdf_data

rdf_data = Parallel(n_jobs=num_cores)(delayed(append_rdf_to_anion)(traj_note = traj_note_dic['traj_note'][i], 
                                                        system = i,
                                                        location = location
                                               )
                           for i, location in enumerate(traj_note_dic['load_location']))

md_data['rdf_anion']  = list(np.concatenate(rdf_data,axis=1)[0])
```

```python
from MDAnalysis.analysis import rdf

def append_solvation_shell_to_cation(traj_note, system, location):
    solv_data = []
    u = mda.Universe(location + '/md.tpr',
                        location + '/md.xtc')
    
    if traj_note == 'CAM':
        cation = u.select_atoms('resname ' + traj_note + ' and type Dz')
    elif traj_note == 'CAL_old':
        cation = u.select_atoms('resname CAL')
    else:
        cation = u.select_atoms('resname ' + traj_note)
    water_list = []
    for ion in cation:
        if traj_note == 'CAL_old':
            traj_note = 'CAL'
        water_list.append(u.select_atoms('name OH2 and around ' + str(first_hydration_shell_dic[traj_note] + 1) + ' index ' + str(ion.index),
                               updating=True))

    for ts in u.trajectory[:md_data[md_data.system==system]['frame'].max() + 1]:
        solv_data.append([water.n_atoms for water in water_list])
    
    return np.asarray(solv_data)

fss_data = Parallel(n_jobs=num_cores)(delayed(append_solvation_shell_to_cation)(traj_note = traj_note_dic['traj_note'][i], 
                                                        system = i,
                                                        location = location
                                               )
                           for i, location in enumerate(traj_note_dic['load_location']))


md_data['solvation_shell_cation'] = [data  for data_sys in fss_data
                                            for data in data_sys]
```

```python
md_data['solvation_shell_cation_mean'] = md_data['solvation_shell_cation'].apply(np.mean)
md_data['solvation_shell_cation_std'] = md_data['solvation_shell_cation'].apply(np.std)
```

```python
from MDAnalysis.analysis import rdf

def append_solvation_shell_to_anion(traj_note, system, location):
    solv_data = []
    u = mda.Universe(location + '/md.tpr',
                        location + '/md.xtc')
    
    anion = u.select_atoms('resname CLA')
    water_list = []
    for ion in anion:
        water_list.append(u.select_atoms('name OH2 and around ' + str(first_hydration_shell_dic['CLA'] + 1) + ' index ' + str(ion.index),
                               updating=True))

    for ts in u.trajectory[:md_data[md_data.system==system]['frame'].max() + 1]:
        solv_data.append([water.n_atoms for water in water_list])
    
    return np.asarray(solv_data)

fss_data = Parallel(n_jobs=num_cores)(delayed(append_solvation_shell_to_anion)(traj_note = traj_note_dic['traj_note'][i], 
                                                        system = i,
                                                        location = location
                                               )
                           for i, location in enumerate(traj_note_dic['load_location']))


md_data['solvation_shell_anion'] = [data for data_sys in fss_data
                                            for data in data_sys]
```

```python
md_data['solvation_shell_anion_mean'] = md_data['solvation_shell_anion'].apply(np.mean)
md_data['solvation_shell_anion_std'] = md_data['solvation_shell_anion'].apply(np.std)
```
