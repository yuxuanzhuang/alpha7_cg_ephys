```python
import matplotlib.pyplot as plt
import matplotlib as mpl

colors = [palette_per[0], (1, 1, 1), (1,0,0)]
cmap_name = 'closed'
cmap = LinearSegmentedColormap.from_list(cmap_name, colors, N=100)

fig, ax = plt.subplots(figsize=(1,10))
fig.subplots_adjust(bottom=0.5)

norm = mpl.colors.Normalize(vmin=0, vmax=6)

cb1 = mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                norm=norm,
                                orientation='vertical')
cb1.set_label('Residence Time (Î¼s)')
plt.savefig('Figures/CG/residence_time_closed_cbar.pdf',bbox_inches = 'tight', pad_inches=0.1, transparent=False)
```
