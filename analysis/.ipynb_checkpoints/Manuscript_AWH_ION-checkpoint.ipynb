```python
class AWH_XVG(object):
    def __init__(self, time, location):
        self.filename = location
        self.md_name = location.split('/')[-3]

        if location.split('/')[-3] == 'PNU_EPJ_0CAL':
            self.ion_type = "wo_b_ca"
        else:
            self.ion_type = location.split('/')[-2].split('_')[-1]
        self.time = time
        self.read()
        
    def read(self):
        self.pmf = pd.read_table(self.filename,
                                 header=None,
                                 names=['coord','PMF','coord_bias','coord_distr','ref_value_distr','target_ref_value_distr','friction_metric'],
                                 sep='\s+',
                                 error_bad_lines=False,
                                 skiprows=31)
        self.awh_target_error = eval(linecache.getline(self.filename, 30).split(' ')[-2])
        self.log_sample_weight = eval(linecache.getline(self.filename, 31).split(' ')[-1])
        self.pmf = self.pmf.applymap(float)
        self.pmf.coord = - self.pmf.coord * 10
        self.pmf.PMF = self.pmf.PMF - np.mean(self.pmf.PMF[:10])
        self.pmf.PMF = self.pmf.PMF / 4.184
```

```python
class PULLX_XVG(object):
    def __init__(self, walker, location):
        self.filename = location + '/walker' + str(walker) + '/awh_pullx.xvg'
        self.md_name = location.split('/')[-3]

        self.ion_type = location.split('/')[-2].split('_')[-1]
        self.walker = walker
        self.read()
        
    def read(self):
        self.pullx = pd.read_table(self.filename,
                                 header=None,
                                 names=['Time','Position'],
                                 sep='\s+',
                                 error_bad_lines=False,
                                 skiprows=17)
        self.pullx = self.pullx.iloc[::100, :]
        self.pullx.Position = self.pullx.Position.apply(float)
        self.pullx.Time = self.pullx.Time.apply(float)

        self.pullx.Position = - self.pullx.Position * 10
        self.pullx.Time = self.pullx.Time / 1000
```

```python
def append_metadata(awh_location):
    rep_data = []
    awh_files = []
    for f in glob.glob(awh_location + '/awh*'):
        awh_files.append(f)
    if awh_files == []:
        print(awh_location, 'empty')
        
    if awh_location == '../trajectory/PNU_EPJ/AWH_CAM_0CAL':
        md_name = 'PNU_EPJ_0CAL'
        ion_type = 'CAM'
        awh_files.sort(key=lambda x: eval(x.split('/')[-1].split('.')[0].split('_')[1][1:])) 
        for i, file in enumerate(awh_files):
            rep_data.append([md_name, eval(file.split('/')[-1].split('.')[0].split('_')[1][1:]), ion_type,file])   

        return rep_data
    
    if awh_location == '../trajectory/PNU_EPJ/AWH_SOD_0CAL':
        md_name = 'PNU_EPJ_0CAL'
        ion_type = 'SOD'
        awh_files.sort(key=lambda x: eval(x.split('/')[-1].split('.')[0].split('_')[1][1:])) 
        for i, file in enumerate(awh_files):
            rep_data.append([md_name, eval(file.split('/')[-1].split('.')[0].split('_')[1][1:]), ion_type,file])   

        return rep_data
    
    if awh_location == '../trajectory/PNU_EPJ/AWH_CLA_0CAL':
        md_name = 'PNU_EPJ_0CAL'
        ion_type = 'CLA'
        awh_files.sort(key=lambda x: eval(x.split('/')[-1].split('.')[0].split('_')[1][1:])) 
        for i, file in enumerate(awh_files):
            rep_data.append([md_name, eval(file.split('/')[-1].split('.')[0].split('_')[1][1:]), ion_type,file])   

        return rep_data
    
    md_name = awh_location.split('/')[-2]
    ion_type = awh_location.split('/')[-1].split('_')[-1]
    awh_files.sort(key=lambda x: eval(x.split('/')[-1].split('.')[0].split('_')[1][1:])) 
    for i, file in enumerate(awh_files):
        rep_data.append([md_name, eval(file.split('/')[-1].split('.')[0].split('_')[1][1:]), ion_type,file])   

    return rep_data
    
    
meta_data = Parallel(n_jobs=num_cores)(delayed(append_metadata)(awh_location
                                               )
                           for awh_location in awh_dict)

awh_data = pd.DataFrame(columns=list(['MD_name','traj_time','ion','file_loc']))
for i in range(0, len(awh_dict)):
    awh_data = awh_data.append(pd.DataFrame(meta_data[i],columns=list(['MD_name','traj_time','ion','file_loc'])),ignore_index=True)
awh_data['traj_time'] =awh_data['traj_time'].apply(int)
```

```python
def append_awh_data(row):
    return AWH_XVG(row.traj_time, row.file_loc)

awh_files = Parallel(n_jobs=num_cores)(delayed(append_awh_data)(row)
                           for i, row in awh_data.iterrows())

awh_data['awh_data'] = awh_files
```

```python
def plot_awh_pmf(awh_file, xdata, ax, skip=10):
    l = ax.plot(awh_file.pmf[xdata],
                awh_file.pmf['coord'],markersize=0)
    ax.set_ylim(-100,100)
#    ax.set_xlabel(xdata + ' (kcal/mol)')
    ax.set_xlabel('Energy' + ' (kcal/mol)',fontsize=25)

    ax.set_ylabel('Pore axis(Å)',fontsize=25)
    l[0].set_linewidth(2)
    if awh_file.ion_type == 'SOD':
        marker = "*"
        ion_lab = 'Sodium'
        
    if awh_file.ion_type == 'CAL':
        marker = "h"
        ion_lab = 'Calcium CHARMM'

    if awh_file.ion_type == 'CAM':
        marker = "+"
        ion_lab = 'Calcium m-site'

    if awh_file.ion_type == 'CLA':
        marker = "v"
        ion_lab = 'Chloride'

    if awh_file.ion_type == 'POT':
        marker = "D"
        ion_lab = 'Potassium'
        
    if awh_file.ion_type == '0CAL':
        marker = "x"
        ion_lab = 'w/o bound Ca'
        
    if awh_file.ion_type == 'w_b_ca':
        marker = "x"
        ion_lab = 'With Ca'
        
    if awh_file.ion_type == 'wo_b_ca':
        marker = "x"
        ion_lab = 'w/o bound Ca'

    sca = ax.scatter(awh_file.pmf[xdata][::skip],
                     awh_file.pmf['coord'][::skip],
                     marker = marker,
                     s=50,
#                     c=palette[awh_file.md_name],
                     c=ion_palette[awh_file.ion_type],
                     linewidths = 3,
                     label=ion_lab,
                     edgecolors=None)
    
#    l[0].set_color(palette[awh_file.md_name])
    l[0].set_color(ion_palette[awh_file.ion_type])

    return l, sca
```

```python
def plot_auxilary(ax, ion=True, system=True):
    ax2 = ax.twinx()  # instantiate a second axes that shares the same x-axis

    ax2.set_ylabel('Annotations')  # we already handled the x-label with ax1
    ax2.tick_params(axis='y',direction='in')
    ax2.set_ylim(-100,100)

    ax2.spines['top'].set_linewidth(4)
    ax2.spines['right'].set_linewidth(4)
    ax2.yaxis.set_tick_params(width=4)
    ax2.set_yticks([95,35,0,-35,-75])
    ax2.set_yticklabels(['SOL','ECD','TMD','ICD','SOL'])

    if ion:
        bar_list = []
        bar_list.append(plt.Line2D([0], [0],color='black',linewidth=1,marker='*',markerfacecolor='white'))
        bar_list.append(plt.Line2D([0], [0],color='black',linewidth=1,marker='h',markerfacecolor='white'))
        bar_list.append(plt.Line2D([0], [0],color='black',linewidth=1,marker='x',markerfacecolor='white'))
        bar_list.append(plt.Line2D([0], [0],color='black',linewidth=1,marker='v',markerfacecolor='white'))
        bar_list.append(plt.Line2D([0], [0],color='black',linewidth=1,marker='D',markerfacecolor='white'))
        ion_list = ['SOD','CAL','CAM','CLA','POT']
        l = plt.legend(bar_list, ion_list, loc=3, ncol = 1, prop={'size':15})
        l.draw_frame(False)

        ax = plt.gca().add_artist(l)
    if system: 
        bar_list_2 = []
        bar_list_2.append(plt.Line2D([0], [0],color='#e66101',linewidth=3))
        bar_list_2.append(plt.Line2D([0], [0],color='#fdb863',linewidth=3))
        bar_list_2.append(plt.Line2D([0], [0],color='#5e3c99',linewidth=3))
        sim_list = ['EPJ_PNU','EPJ','BGT']
        l_2 = plt.legend(bar_list_2, sim_list, loc=1, ncol = 3, prop={'size':15})
        l_2.draw_frame(False)
```

```python
def plot_radius_hf(ax, protein_list=None, radius=True, hf=True):

    pore_radius_data = pd.read_pickle('radius_structure.pickle')
    pore_radius_data = pore_radius_data[pore_radius_data.selection=='protein']
    
    ax3 = ax.twiny()
    ax3.set_xlabel('Radius (Å) / Hydrophobicity')
    
    if protein_list is None:
        protein_list = set(pore_radius_data.protein)

    if radius:

        for ((protein), radius), color in zip(pore_radius_data.groupby(['protein'],sort=False), ['#e66101', '#fdb863', '#5e3c99']):
            if protein in protein_list:
                ax3.plot(radius.radius*10,
                         radius.s*10,
                         color=color,
                         linewidth=1)


    if hf: 
        for ((protein), radius), color in zip(pore_radius_data.groupby(['protein'],sort=False), ['#e66101', '#fdb863', '#5e3c99']):
            if protein in protein_list:
                ax3.scatter(radius.pl_hydrophobicity[::10]*10,
                            radius.s[::10]*10,
                            color=color,
                            s=1)
                
    ax3.set_xlim(ax3.get_xlim()[::-1])
```

```python
def create_md_dataframe():    
    return pd.DataFrame(columns=list(['MD_name','ion','walker','frame','traj_time','system','ligand','note']))
```

```python
def append_metadata(traj_note, ion, walker, system, location):
        rep_data = []
        traj_note_split = traj_note.split('_')
        try:
            top_location = 'init.pdb'
            traj_location = 'walker' + walker + '/awh.xtc'
            traj = mda.Universe(location + '/AWH_' + ion + '/' + top_location,
                                location + '/AWH_' + ion + '/' + traj_location)         
            md_name = traj_note_split[0]
            ligand = traj_note_split[1:-1] ##in this case
            note = traj_note_split[-1]
            n_frames = traj.trajectory.n_frames
            ts = traj.trajectory.dt
            for i in range(n_frames):
                rep_data.append([md_name, ion, walker, i,  i * ts, system, ligand, note])

        except:
            print(traj_note + ion + ' not found.')
        return rep_data
    
    
meta_data = Parallel(n_jobs=num_cores)(delayed(append_metadata)(traj_note = traj_note, 
                                                        ion = ion,
                                                        walker = str(walker),
                                                        system = ind,
                                                        location = traj_note_dic['save_location'][ind],
                                               )
                           for ind, traj_note in enumerate(traj_note_dic['traj_note'])
                           for walker in range(1,traj_note_dic['walker'][ind] + 1)
                           for ion in traj_note_dic['ion_type']
                           )
md_data = create_md_dataframe()
for i in range(0, len(meta_data)):
    md_data = md_data.append(pd.DataFrame(meta_data[i],columns=list(['MD_name','ion','walker','frame','traj_time','system','ligand','note'])),ignore_index=True)
md_data['frame'] = md_data['frame'].apply(int)
md_data['traj_time'] = md_data['traj_time'].apply(float)
md_data['walker'] = md_data['walker'].apply(int)
md_data['system'] = md_data['system'].apply(int)
```

```python
def append_pore_ion_info(traj_note, ion, walker, location):
    def cartesian_to_cylinder(posx, posy, princ_x, princ_y):
        return np.sqrt((posx - princ_x) ** 2 + (posy - princ_y) ** 2)
    
    poreion_data = []
    traj_note_split = traj_note.split('_')
    top_location = 'init.pdb'
    traj_location = 'walker' + walker + '/awh.xtc'
    traj = mda.Universe(location + '/AWH_' + ion + '/' + top_location,
                        location + '/AWH_' + ion + '/' + traj_location)
    
    ion = ion.split('_')[0]
    if ion == 'CAM':
        pore_ion = traj.atoms[-7]
        pore_ion_index = traj.atoms.n_atoms - 7
        solvation_shell = traj.select_atoms('name OH2 and around ' + str(first_hydration_shell_dic[ion] + 1) + ' index ' + str(pore_ion_index), updating=True)
        protein_shell = traj.select_atoms('protein and type O and around ' + str(first_hydration_shell_dic[ion] + 1) + ' index ' + str(pore_ion_index), updating=True)

    else:
        pore_ion = traj.atoms[-1]
        pore_ion_index = traj.atoms.n_atoms - 1
        solvation_shell = traj.select_atoms('name OH2 and around '+ str(first_hydration_shell_dic[ion] + 1) + ' index ' + str(pore_ion_index), updating=True)
        protein_shell = traj.select_atoms('protein and type O and around ' + str(first_hydration_shell_dic[ion] + 1) + ' index ' + str(pore_ion_index), updating=True)

    princ_center = np.mean(traj.select_atoms('resid 247 and name CA').positions[:5],axis=0)

    for ts in traj.trajectory:
        position_x = pore_ion.position.T[0]
        position_y = pore_ion.position.T[1]
        position_z = pore_ion.position.T[2] - princ_center[2]
        position_r = cartesian_to_cylinder(position_x, position_y, princ_center[0], princ_center[1])

        poreion_data.append(np.asarray([position_x,
                                        position_y,
                                        position_z,
                                        position_r,
                                        solvation_shell.n_atoms,
                                        protein_shell.n_atoms,
                                        protein_shell.resids]).T)
    return poreion_data
    
ion_data = Parallel(n_jobs=num_cores)(delayed(append_pore_ion_info)(traj_note = traj_note, 
                                                        ion = ion,
                                                        walker = str(walker),
                                                        location = traj_note_dic['save_location'][ind],
                                               )
                           for ind, traj_note in enumerate(traj_note_dic['traj_note'])
                           for walker in range(1,traj_note_dic['walker'][ind] + 1)
                           for ion in traj_note_dic['ion_type']
                           )

ion_data_concat = [x for x in ion_data if x != []]
md_data['pore_ion_data'] = list(np.concatenate(ion_data_concat,axis=0))
```

```python
md_data['pore_ion_pos_x'] = md_data['pore_ion_data'].apply(lambda x: x[0])
md_data['pore_ion_pos_y'] = md_data['pore_ion_data'].apply(lambda x: x[1])
md_data['pore_ion_pos_z'] = md_data['pore_ion_data'].apply(lambda x: x[2])
md_data['pore_ion_pos_r'] = md_data['pore_ion_data'].apply(lambda x: x[3])
md_data['pore_ion_hydration_shell'] = md_data['pore_ion_data'].apply(lambda x: x[4])
md_data['pore_ion_protein_shell'] = md_data['pore_ion_data'].apply(lambda x: x[5])
md_data['pore_ion_protein_shell_resid'] = md_data['pore_ion_data'].apply(lambda x: x[6])
```

```python
from gridData import Grid
def get_plot_potential(file, protein):

    emap = Grid(file + '/electrostatic_potential.dx')
    wmap = Grid(file + '/water_occupancy.dx')

    threshold = 0.1

    wslice = wmap.grid[:,:,-1]
    eslice = emap.grid[:,:,-1]
    p0 = np.mean(eslice[wslice>threshold])

    ele_data = []
    for k in range(emap.grid.shape[2]-1, -1, -1):
        if k > 80:
            radius = 7
        else:
            radius = 19

        z = emap.origin[2] + k*emap.delta[2]
        wslice = wmap.grid[19 - radius:19 + radius,
                           19 - radius:19 + radius,
                           k]
        eslice = emap.grid[19 - radius:19 + radius,
                           19 - radius:19 + radius,
                           k]

        p = np.mean(eslice[wslice>threshold])
        ele_data.append([z, p-p0])

    ele_data = np.asarray(ele_data)

#    l = ax.plot(ele_data.T[1],ele_data.T[0], label='Open', linewidth=4)
    
    u = mda.Universe(protein)
    data = {}
    data['axis'] = ele_data.T[0] - u.select_atoms('resid 269 and name CA').positions.T[2].mean() + (u.select_atoms('resid 269 and name CA').positions.T[2].mean() - u.select_atoms('resid 247 and name CA').positions.T[2].mean())
    data['potential'] = ele_data.T[1]

    fig, ax = plt.subplots(figsize=(5,10), gridspec_kw={'wspace': 0})

    l = ax.plot(data['potential'], data['axis'], label='Open', linewidth=4)

    #ax.set_xlim(-1,10)
    #ax.set_xticks(np.arange(0,9))
    ax.set_ylim(-75,95)

    for ind, domain in enumerate(domain_annotations_plot.items()):
        ax.axhspan(domain[1][0], domain[1][-1], facecolor=two_shades[0], alpha=0.1)

    for ind, residue in enumerate(residue_annotations_plot.items()):
        ax.axhline(residue[1], c=two_shades[1], linewidth=2, linestyle='--',alpha=1)

    set_axis_boarder(ax)

    ax.legend(loc=3)
    ax.set_xlabel('Potential (mV)')
    ax.set_ylabel('Pore Axis (Å)')
    return data
```
