```python
def create_md_dataframe():    
    return pd.DataFrame(columns=list(['MD_name','replicate','frame','traj_time','system','id','ligand','note']))
```

```python
def append_metadata(traj_note, rep, ident, system, location, skip = default_skip):
        rep_data = []
        traj_note_split = traj_note.split('_')
        try:
            top_location =  "/eq.pdb"
            traj_location = '/multijob/rep' + rep + '/' + "cg.xtc"
            traj = mda.Universe(location + top_location,
                                location + traj_location)         
            md_name = traj_note_split[0]
            ligand = traj_note_split[1:-1] ##in this case
            timestep = 2
            note = traj_note_split[-1]
            n_frames = traj.trajectory.n_frames
            ts = traj.trajectory.dt
            for i in range(0, n_frames,10):
                rep_data.append([md_name, rep, i,  ts * i, system, ident,ligand,note])
                
        except:
            print(traj_note + ' not found.')
        return rep_data
    
    
meta_data = Parallel(n_jobs=num_cores)(delayed(append_metadata)(traj_note = traj_note_dic['traj_note'][i//4], 
                                                        rep = str(i%4 + 1),
                                                        ident = i,
                                                        system = i//4,
                                                        location = traj_note_dic['save_location'][i//4],
                                                        skip = traj_note_dic['skip'][i//4], 
                                               )
                           for i in range(0, np.sum(traj_note_dic['rep'])))
md_data = create_md_dataframe()
for i in range(0,np.sum(traj_note_dic['rep'])):
    md_data = md_data.append(pd.DataFrame(meta_data[i],columns=list(['MD_name','replicate','frame','traj_time','system','id','ligand','note'])),ignore_index=True)
md_data['frame'] =md_data['frame'].apply(int)
md_data['traj_time'] =md_data['traj_time'].apply(float)
md_data['replicate'] =md_data['replicate'].apply(int)
md_data['system'] = md_data['system'].apply(int)
```

```python
def append_pnu_info(traj_note, system, rep, location):
    def cartesian_to_cylinder_r(posx, posy, princ_x, princ_y):
        return np.sqrt((posx - princ_x) ** 2 + (posy - princ_y) ** 2)
    def cartesian_to_cylinder_theta(posx, posy, princ_x, princ_y):
        return np.arctan2((posy - princ_y) , (posx - princ_x))
    
    lipid_data = []
    traj_note_split = traj_note.split('_')
    top_location =  "/eq.pdb"
    traj_location = '/multijob/rep' + rep + '/' + "cg.xtc"
    traj = mda.Universe(location + top_location,
                                location + traj_location)     
    lipid = traj.select_atoms('resname pnu')


    princ_center_ch = np.mean(traj.select_atoms('resid 247 and name BB').positions,axis=0)

#    for ts in traj.trajectory[:md_data[(md_data.system==system) & (md_data.replicate == eval(rep))]['frame'].max()+1]:
    for ts in traj.trajectory[::10]:
        position_x = lipid.positions.T[0]
        position_y = lipid.positions.T[1]
        position_z = lipid.positions.T[2] - princ_center_ch[2]
        position_r = cartesian_to_cylinder_r(position_x, position_y, princ_center_ch[0], princ_center_ch[1])
        position_theta = cartesian_to_cylinder_theta(position_x, position_y, princ_center_ch[0], princ_center_ch[1])

        lipid_data.append(np.asarray([position_x,
                                    position_y,
                                    position_z,
                                    position_r,
                                    position_theta / np.pi * 180]).T)
    return lipid_data


pnu_data = Parallel(n_jobs=num_cores)(delayed(append_pnu_info)(
                                                        traj_note = traj_note_dic['traj_note'][i//4], 
                                                        system = i//4,
                                                        rep = str(i%4 + 1),
                                                        location = traj_note_dic['save_location'][i//4],
                                               )
                           for i in range(0, np.sum(traj_note_dic['rep'])))

pnu_data_concat = [x for x in pnu_data if x != []]
md_data['pnu_data'] = [item for sublist in pnu_data_concat for item in sublist]
```

```python
from MDAnalysis.analysis.rms import rmsd
def append_pnu_rmsd(traj_note, system, rep, location):
    lipid_data = []
    traj_note_split = traj_note.split('_')
    top_location =  "/eq.pdb"
    traj_location = '/multijob/rep' + rep + '/' + "cg.xtc"
    traj = mda.Universe(location + top_location,
                                location + traj_location)     
    pnus = traj.select_atoms('resname pnu')
    
    pnu_config = mda.Universe('../CG/NACHRS7_PNU_EPJ_CHOL/pnu.pdb')

#    for ts in traj.trajectory[:md_data[(md_data.system==system) & (md_data.replicate == eval(rep))]['frame'].max()+1]:
    for ts in traj.trajectory[::10]:
        rmsd_result = []
        for pnu_single in pnus.residues:
            rmsd_result.append(rmsd(pnu_single.atoms.positions, pnu_config.atoms.positions, superposition = True))

        lipid_data.append(np.asarray(rmsd_result).T)
    return lipid_data


pnu_data = Parallel(n_jobs=num_cores)(delayed(append_pnu_rmsd)(
                                                        traj_note = traj_note_dic['traj_note'][i//4], 
                                                        system = i//4,
                                                        rep = str(i%4 + 1),
                                                        location = traj_note_dic['save_location'][i//4],
                                               )
                           for i in range(0, np.sum(traj_note_dic['rep'])))

pnu_data_concat = [x for x in pnu_data if x != []]
md_data['pnu_rmsd_data'] = [item for sublist in pnu_data_concat for item in sublist]
```

```python
def append_metadata(traj_note, rep, ident, system, location, skip = default_skip):
        rep_data = []
        traj_note_split = traj_note.split('_')
        try:
            top_location =  "/eq.pdb"
            traj_location = '/multijob/rep' + rep + '/' + "cg.xtc"
            traj = mda.Universe(location + top_location,
                                location + traj_location)         
            md_name = traj_note_split[0]
            ligand = traj_note_split[1:-1] ##in this case
            timestep = 2
            note = traj_note_split[-1]
            n_frames = traj.trajectory.n_frames
            ts = traj.trajectory.dt
            for i in range(0, n_frames,10):
                for pnu_ind in range(10):
                    rep_data.append([md_name, rep, i,  ts * i, system, ident, pnu_ind, note])
                
        except:
            print(traj_note + ' not found.')
        return rep_data
    
    
meta_data = Parallel(n_jobs=num_cores)(delayed(append_metadata)(traj_note = traj_note_dic['traj_note'][i//4], 
                                                        rep = str(i%4 + 1),
                                                        ident = i,
                                                        system = i//4,
                                                        location = traj_note_dic['save_location'][i//4],
                                                        skip = traj_note_dic['skip'][i//4], 
                                               )
                           for i in range(0, np.sum(traj_note_dic['rep'])))
pnu_data = create_md_dataframe()
for i in range(0,np.sum(traj_note_dic['rep'])):
    pnu_data = pnu_data.append(pd.DataFrame(meta_data[i],columns=list(['MD_name','replicate','frame','traj_time','system','id','ligand','note'])),ignore_index=True)
pnu_data['frame'] =pnu_data['frame'].apply(int)
pnu_data['traj_time'] =pnu_data['traj_time'].apply(float)
pnu_data['replicate'] =pnu_data['replicate'].apply(int)
pnu_data['system'] = pnu_data['system'].apply(int)
pnu_data['ligand'] = pnu_data['ligand'].apply(int)
```
